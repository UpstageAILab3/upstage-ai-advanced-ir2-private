```python
import pandas as pd
evaljsonl = pd.read_json('/data/ephemeral/home/data/eval.jsonl', lines=True)
highscore = pd.read_json('/data/ephemeral/home/data/highscore.csv', lines=True)
```

```python
# 각 eval_id당 모든 content를 하나의 문장으로 이어 저장할 리스트 초기화
eval_contents = []

# 각 행의 'msg' 리스트에서 'content'를 하나의 문장으로 이어붙임
for index, row in evaljsonl.iterrows():
    if 'msg' in row:
        combined_content = ' '.join([message['content'] for message in row['msg'] if 'content' in message])
        eval_contents.append({'eval_id': row['eval_id'], 'combined_content': combined_content})

# 모든 eval_id와 결합된 content 출력
for item in eval_contents:
    print(f"eval_id: {item['eval_id']}, combined_content: {item['combined_content']}")
```

```python
# 모든 eval_id와 결합된 content를 새로운 JSONL 파일로 저장
import json
with open('/data/ephemeral/home/data/eval_combined.jsonl', 'w') as f:
    for item in eval_contents:
        f.write(f"{json.dumps(item, ensure_ascii=False)}\n")
```

```python
# eval_combined 데이터 불러오기
eval_combined = pd.read_json('/data/ephemeral/home/data/eval_combined.jsonl', lines=True)

# eval_id를 기준으로 highscore에 combined_content 추가
highscore = highscore.merge(eval_combined, on='eval_id', how='left')

# 업데이트된 highscore를 저장
highscore.to_json('/data/ephemeral/home/data/highscore_with_combined_content.jsonl', orient='records', lines=True, force_ascii=False)
```

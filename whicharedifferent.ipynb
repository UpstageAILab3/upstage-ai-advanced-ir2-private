```python
import pandas as pd
highscore = pd.read_json('/data/ephemeral/home/data/highscore.csv', lines=True)
output = pd.read_json('/data/ephemeral/home/sample_submission14_roberta_hybridmodified_synonyms_reranking_jimiprompt_documeta.csv', lines=True)
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows',None) 
```

```python
# topk 첫번째만 비교
# highscore와 output을 eval_id를 기준으로 merge
merged_df = pd.merge(highscore, output, on='eval_id', suffixes=('_highscore', '_output'))

# topk의 첫 번째 값 비교 (리스트의 첫 번째 요소만 비교, 문자열 처리 추가)
def get_first_element(value):
    if isinstance(value, list) and len(value) > 0:
        return value[0]
    elif isinstance(value, str):
        return value.split(',')[0].strip()
    return value

merged_df['topk_highscore_first'] = merged_df['topk_highscore'].apply(get_first_element)
merged_df['topk_output_first'] = merged_df['topk_output'].apply(get_first_element)
merged_df['same_top1'] = merged_df['topk_highscore_first'] == merged_df['topk_output_first']

# 결과 출력
result_df = merged_df[['eval_id', 'same_top1']]
print(result_df)
```

```python
# 세 개를 각각 비교교
# highscore와 output을 eval_id를 기준으로 merge
merged_df = pd.merge(highscore, output, on='eval_id', suffixes=('_highscore', '_output'))

# topk의 첫 번째, 두 번째, 세 번째 값 추출 (리스트의 요소만 비교, 문자열 처리 추가)
def get_elements(value, n):
    if isinstance(value, list) and len(value) >= n:
        return value[:n]
    elif isinstance(value, str):
        elements = value.split(',')
        return [elem.strip() for elem in elements[:n]]
    return []

merged_df['topk_highscore_elements'] = merged_df['topk_highscore'].apply(lambda x: get_elements(x, 3))
merged_df['topk_output_elements'] = merged_df['topk_output'].apply(lambda x: get_elements(x, 3))

# 첫 번째, 두 번째, 세 번째 값 각각 비교
merged_df['same_top1'] = merged_df.apply(lambda row: row['topk_highscore_elements'][0] == row['topk_output_elements'][0] if len(row['topk_highscore_elements']) > 0 and len(row['topk_output_elements']) > 0 else False, axis=1)
merged_df['same_top2'] = merged_df.apply(lambda row: row['topk_highscore_elements'][1] == row['topk_output_elements'][1] if len(row['topk_highscore_elements']) > 1 and len(row['topk_output_elements']) > 1 else False, axis=1)
merged_df['same_top3'] = merged_df.apply(lambda row: row['topk_highscore_elements'][2] == row['topk_output_elements'][2] if len(row['topk_highscore_elements']) > 2 and len(row['topk_output_elements']) > 2 else False, axis=1)

# 결과 출력
result_df = merged_df[['eval_id', 'same_top1', 'same_top2', 'same_top3']]
print(result_df)
```

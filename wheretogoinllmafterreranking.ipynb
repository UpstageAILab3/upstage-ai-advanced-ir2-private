```python
import pandas as pd
reranking20 = pd.read_json('/data/ephemeral/home/firstreranking.csv', lines=True)
llmtop3 = pd.read_json('/data/ephemeral/home/sample_submission14_snunlpKLUE_combined300_synonyms_reranking(multi_query)_jimiprompt_documeta100_llmcall_rerankertwice.csv', lines=True)
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows',None) 
```

```python
# Merging reranking20 and llmtop3 based on eval_id
merged_df = pd.merge(reranking20, llmtop3, on='eval_id', suffixes=('_reranking20', '_llmtop3'))

# Compare reranking20 and llmtop3 positions after merging
comparison_results = []

for _, row in merged_df.iterrows():
    reranking20_topk = row['topk_reranking20']
    llmtop3_topk = row['topk_llmtop3']

    # Assigning ranks (1 to len(reranking20_topk)) to each ID in reranking20 topk
    reranking20_mapping = {doc_id: idx + 1 for idx, doc_id in enumerate(reranking20_topk)}

    # Mapping llmtop3 documents to their positions in reranking20
    llmtop3_positions = {doc_id: reranking20_mapping.get(doc_id, None) for doc_id in llmtop3_topk}
    
    comparison_results.append({
        "eval_id": row['eval_id'],
        "llmtop3_positions": llmtop3_positions
    })

# Output the results
comparison_results = pd.DataFrame(comparison_results)
comparison_results
```
